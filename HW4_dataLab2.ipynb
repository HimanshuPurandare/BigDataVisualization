```python
import matplotlib.pyplot as plt
import google.datalab.bigquery as bq
import pandas as p

Birth_Count = bq.Query('SELECT CAST(source_year AS string) AS year, COUNT(is_male) AS birth_count FROM `publicdata.samples.natality` WHERE month=12 AND day=02 GROUP BY year ORDER BY year DESC')
df = Birth_Count.execute(output_options=bq.QueryOutput.dataframe()).result()
list1=[]
list1=list(df.head(100).year)

df.plot.pie(labels=list1,x='year',y='birth_count', title = 'birthdays in same month and year',autopct='%.1f',figsize=(15, 15))
```
